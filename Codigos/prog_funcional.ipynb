# 1 Energía Cinética

```python
from functools import reduce

def energia_cinetica(m, v):
    return list(map(lambda m_v: 0.5 * m_v[0] * m_v[1]**2, zip(m, v)))
```

# 2. Números Primos

```python
def es_primo(n):
    return n > 1 and all(n % i != 0 for i in range(2, int(n**0.5) + 1))

def filtrar_primos(numeros):
    return list(filter(es_primo, numeros))
```

# 3. Transformación de Temperatura

```python
def transformar_temp(temp, escala):
    return [t + 273.15 if escala == 'K' else 1.8 * t + 32 for t in temp]
```

# 4. Maxwell-Boltzman

```python
import math

def maxwell_boltzmann(T, m, v):
    k = 1.38e-23
    factor = (m / (2 * math.pi * k * T))**1.5 * 4 * math.pi
    return list(map(lambda v: factor * v**2 * math.exp(-m * v**2 / (2 * k * T)), v))
```

# 5. Carga y Potencial

```python
def potencial_electrico(p, cargas):
    k = 9e9
    return sum(map(lambda pos_q: k * pos_q[1] / ((p[0] - pos_q[0][0])**2 + 
                                                  (p[1] - pos_q[0][1])**2 + 
                                                  (p[2] - pos_q[0][2])**2)**0.5, cargas.items()))
```

# 6. Ondas

```python
def frecuencias_armonicos(cuerdas, v):
    return {L: [n * v / (2 * L) for n in ns] for L, ns in cuerdas.items()}
```

# 7. Método del Trapecio

```python
def integral_trapecio(f, a, b, n):
    h = (b - a) / n
    x_vals = [a + i * h for i in range(n + 1)]
    return (h / 2) * reduce(lambda acc, i: acc + f(x_vals[i]) + f(x_vals[i+1]), range(n), 0)
```

# 8. Gas Ideal

```python
def velocidad_rms(T, M):
    k = 1.38e-23
    return [(3 * k * t / M)**0.5 for t in T]
```

# 9. Euler

```python
def euler(f, x0, y0, h, n):
    return ((x0 + i * h, y0 + sum(map(lambda i: h * f(x0 + i * h, y0), range(i)))) for i in range(n))
```

# 10. Lennard-Jones

```python
def lennard_jones(r, epsilon, sigma):
    return [4 * epsilon * ((sigma / ri)**12 - (sigma / ri)**6) for ri in r]
```
